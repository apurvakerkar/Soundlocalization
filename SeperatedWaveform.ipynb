```python
import wave
import numpy as np
import matplotlib.pyplot as plt

def plot_waveform(wave_file):
    # Open the WAV file
    with wave.open(wave_file, 'rb') as wf:
        # Get the number of channels, sample width (in bytes), and frame rate
        num_channels = wf.getnchannels()
        sample_width = wf.getsampwidth()
        frame_rate = wf.getframerate()
        num_frames = wf.getnframes()

        # Read the audio data
        audio_data = wf.readframes(num_frames)

    # Convert the binary audio data to a numpy array
    audio_array = np.frombuffer(audio_data, dtype=np.int16)

    # Reshape the array to have one column per channel
    audio_array = audio_array.reshape(-1, num_channels)

    # Create a time axis
    time_axis = np.arange(0, num_frames) / frame_rate

    # Plot the waveform for each channel
    for channel in range(num_channels):
        plt.plot(time_axis, audio_array[:, channel], label=f'Channel {channel + 1}')

    plt.xlabel('Time (seconds)')
    plt.ylabel('Amplitude')
    plt.title('Waveform for Each Channel')
    plt.legend()
    plt.show()

# Example usage
wave_file_path = 'C:/Users/apurv/notch_filtered_CF_Chirp.wav'
plot_waveform(wave_file_path)

```

```python
import wave
import numpy as np
import matplotlib.pyplot as plt

def plot_waveform(wave_file):
    # Open the WAV file
    with wave.open(wave_file, 'rb') as wf:
        # Get the number of channels, sample width (in bytes), and frame rate
        num_channels = wf.getnchannels()
        sample_width = wf.getsampwidth()
        frame_rate = wf.getframerate()
        num_frames = wf.getnframes()

        # Read the audio data
        audio_data = wf.readframes(num_frames)

    # Convert the binary audio data to a numpy array
    audio_array = np.frombuffer(audio_data, dtype=np.int16)

    # Reshape the array to have one column per channel
    audio_array = audio_array.reshape(-1, num_channels)

    # Create a time axis
    time_axis = np.arange(0, num_frames) / frame_rate

    # Set the figure size based on the number of channels
    fig, axes = plt.subplots(num_channels, 1, figsize=(8, 2*num_channels))

    # Plot the waveform for each channel in separate subplots
    for channel in range(num_channels):
        axes[channel].plot(time_axis, audio_array[:, channel])
        axes[channel].set_xlabel('Time (seconds)')
        axes[channel].set_ylabel(f'Amplitude (Channel {channel + 1})')

    plt.suptitle('Waveform for Each Channel')
    plt.tight_layout(rect=[0, 0, 1, 0.97])  # Adjust the rect parameter to leave space for suptitle
    plt.show()


# Example usage
wave_file_path = 'C:/Users/apurv/CF_Chirp.wav'
plot_waveform(wave_file_path)
```
